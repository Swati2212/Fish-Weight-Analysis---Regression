### Feature Engineering

```python
# We will perform one hot encoding on the Categorical data ie. 'Species'
a = pd.get_dummies(df['Species'])
```

```python
a
```

```python
df = pd.concat([df,a],axis=1)
```

```python
df.head()
```

```python
df.drop('Species',axis =1,inplace =True)
```

```python
df.head()
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(),annot=True)
```

### Training and Spiliting the Dataset

```python
df = df[['Bream','Parkki','Perch','Pike','Roach','Smelt','Whitefish','Length1','Length2','Length3','Height','Width','Weight']]
df.head()
```

```python
X = df.iloc[:,0:12] # independent features
y = df.iloc[:,12] #dependent features
```

```python
X.shape,y.shape
```

### Preprocessing the dataset

```python
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
print(X_train.shape)
```

```python
print(X_test.shape)
```

```python
print(y_train.shape)
```

```python
print(y_test.shape)
```

### Training Model with Different Algorithms

#### Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
Accuracy = r2_score(y_pred,y_test)
```

```python
print(Accuracy)
```

#### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=120,max_features=12,random_state=0,criterion='mse')
rf.fit(X_train,y_train)
rf_pred = rf.predict(X_test)
Accuracy = r2_score(rf_pred,y_test)
```

```python
print(Accuracy)
```

#### Gradent Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor
gb = GradientBoostingRegressor(n_estimators=120,learning_rate=0.09,random_state=0,max_features=12)
gb.fit(X_train,y_train)
gb_pred = gb.predict(X_test)
Accuracy = r2_score(gb_pred,y_test)
```

```python
print(Accuracy)
```

#### Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(random_state=1,max_features=12)
dt.fit(X_train,y_train)
dt_pred = dt.predict(X_test)
Accuarcy = r2_score(dt_pred,y_test)
```

```python
print(Accuracy)
```

### In all these algorithms, Gradient Boosting Regressor and Decision Tree Regressor have higher accuray. Linear Regression algorithm have the least accuracy among the following.
